## 5. Tu tarea: operar sobre la lista

Completar el siguiente código para:

1. agregar un nuevo movimiento
2. calcular el saldo final
3. contar cuántos cargos y cuántos abonos existen

> Sugerencia:  
> - si el tipo es `"abono"`, el monto suma al saldo  
> - si el tipo es `"cargo"`, el monto resta al saldo

```python
# 1) Agrega un nuevo movimiento
nuevo_movimiento = {
    "descripcion": "Pago internet",
    "monto": 24990,
    "tipo": "cargo"
}
```

```python
# TODO: agregar el nuevo movimiento a la lista
# movimientos.________(nuevo_movimiento)

movimientos.append(nuevo_movimiento)

print("Lista actualizada:")
for m in movimientos:
    print(m)

print("\n" + "="*50)
```

```python
#Calcular el saldo final
saldo = 0

for m in movimientos:
    # Si el tipo es "abono", el monto suma al saldo
    if m["tipo"] == "abono":
        saldo += m["monto"]
    # Si el tipo es "cargo", el monto resta al saldo
    elif m["tipo"] == "cargo":
        saldo -= m["monto"]

print("Saldo final:", saldo)

print("\n" + "="*50)
```

```python
# 3) Cuenta cuántos cargos y cuántos abonos existen
cantidad_cargos = 0
cantidad_abonos = 0

for m in movimientos:
    # Si es cargo se le suma 1 qal cargo
    if m["tipo"] == "cargo":
        cantidad_cargos += 1
    # Si es abono se le sube 1 al abono
    elif m["tipo"] == "abono":
        cantidad_abonos += 1

print("Cantidad de cargos:", cantidad_cargos)
print("Cantidad de abonos:", cantidad_abonos)
```

## 6. Reflexión: del uso de listas al concepto de TAD

Hasta ahora se ha usado una lista de Python directamente.  
Pero en estructura de datos nos interesa pensar más allá del lenguaje.

### Un TAD Lista se define por operaciones como:
- crear una lista vacía
- agregar un elemento
- obtener un elemento por posición
- eliminar un elemento
- conocer su tamaño
- recorrer sus elementos
- verificar si está vacía

### Preguntas
Responder brevemente:

1. ¿Qué parte de esto corresponde a la **abstracción**?
2. ¿Qué parte corresponde a la **implementación**?

```python
abstraccion =  "La abstracción es la parte donde definimos qué operaciones queremos que haga nuestra lista, sin pensar todavía en el código de Python. Por ejemplo, saber que una cuenta bancaria necesita 'agregar un movimiento', 'mostrar el saldo' o 'eliminar un registro' es la abstracción. Es como el manual de usuario: te dice qué hace el botón, pero no cómo funciona el motor por dentro"
implementacion = "La implementación es el código real que escribimos para que esas ideas funcionen. En nuestro caso, es usar una 'lista de diccionarios' de Python, escribir los ciclos for para el saldo o usar el método .append() para guardar los datos. Es cuando aterrizamos la idea abstracta al lenguaje de programación para que la computadora la ejecute"

print("Abstracción:", abstraccion)
print("Implementación:", implementacion)
```

## 8. Tu tarea: completar la implementación

Se debe implementar los siguientes métodos:

- `obtener(posicion)`
- `eliminar(posicion)`
- `mostrar_todos()`
- `calcular_saldo()`

### Recomendaciones
- validar si la posición existe
- decidir cuándo conviene usar `return`
- probar cada método apenas se complete

```python
class ListaMovimientos:
    def __init__(self):
        self.movimientos = []

    def esta_vacia(self):
        return len(self.movimientos) == 0

    def agregar(self, movimiento):
        self.movimientos.append(movimiento)

    def tamano(self):
        return len(self.movimientos)

    def obtener(self, posicion):
        if 0 <= posicion < len(self.movimientos):
            return self.movimientos[posicion]
        return "Error: Posición fuera de rango"

    def eliminar(self, posicion):
        if 0 <= posicion < len(self.movimientos):
            return self.movimientos.pop(posicion)
        return "Error: No se pudo eliminar"

    def mostrar_todos(self):
        print("--- Listado de Movimientos ---")
        for m in self.movimientos:
            print(m)

    def calcular_saldo(self):
        saldo = 0
        for m in self.movimientos:
            if m["tipo"] == "abono":
                saldo += m["monto"]
            else:
                saldo -= m["monto"]
        return saldo

    # --- MÉTODOS DEL DESAFÍO OPCIONAL ---
    def generar_resumen(self):
        print("********** RESUMEN **********")
        print(f"Total Movimientos: {self.tamano()}")
        print(f"Saldo Final: ${self.calcular_saldo()}")
        print("*****************************")

    def buscar_por_descripcion(self, palabra):
        for m in self.movimientos:
            if palabra.lower() in m["descripcion"].lower():
                print(m)

    def filtrar_cargos_mayores_a(self, monto):
        for m in self.movimientos:
            if m["tipo"] == "cargo" and m["monto"] > monto:
                print(m)
```
